## Data Preparation

For purposes of our analysis and more specifically when working with categorical data on machine learning, one-hot encoding is a critical data preparation step. It is required because achine learning models, particularly those implemented in libraries such as scikit-learn, require numerical input data/ One-hot encoding is used to convert categorical information into a binary matrix, where each category is represented by a distinct binary column with 0s and 1s. Because it avoids the requirement to intrepret the original categories as numerical values, which may lead to undesired ordinal associations, the translation assures that the machine learning algorithm can funtion successfully with the data.

```python

# Using pandas' get_dummies to perform one-hot encoding
df = pd.get_dummies(df, columns=['schoolsup','guardian','activities', 'Pstatus', 'famsup', 'paid', 'famsize', 'sex', 'address', 'school'])
df.columns  
```

### Improvements

In this stage, I am investigating how to improve our model's prediction performance through the use of ensemble learning approaches, particularly AdaBoost. The predictions from two basic models, Support Vector Regression (SVR) and Linear Regression, have been integrated, and AdaBoost is now being applied to this ensemble to increase the predictive potential of our model even more.
Because of AdaBoostRegressor's versatility, capacity to raise prediction accuracy, and ability to optimize and balance the combined predictions of SVR and Linear Regression, I selected it as our ensemble technique. With this combination, I am able to develop a reliable and precise prediction model for our dataset.
The goal of ensemble approaches, like AdaBoost, is to produce a more powerful and reliable prediction model by aggregating the results of several basic models. The foundation models in this instance are SVR and linear regression, each of which has advantages and disadvantages. In order to maximize the specific strengths of each of these models while minimizing their shortcomings, I plan to use AdaBoost after averaging their forecasts.

```python

linear_reg = LinearRegression(fit_intercept=False, positive=True)
svr = SVR(C=10, epsilon=1, kernel='linear')
linear_reg.fit(X_train, y_train)
svr.fit(X_train, y_train)

linear_reg_preds = linear_reg.predict(X_test)
svr_preds = svr.predict(X_test)

ensemble_preds = (linear_reg_preds + svr_preds) / 2

base_estimator = DecisionTreeRegressor(max_depth=3)
base_estimator.fit(ensemble_preds.reshape(-1, 1), y_test)

adaboost = AdaBoostRegressor(base_estimator=base_estimator, n_estimators=50, learning_rate=1.0, random_state=42)

adaboost.fit(ensemble_preds.reshape(-1, 1), y_test)
joblib.dump(linear_reg, 'your_linear_reg_model.joblib')
joblib.dump(svr, 'your_svr_model.joblib')
joblib.dump(adaboost, 'your_adaboost_model.joblib')

linear_reg_preds_test = linear_reg.predict(X_test)
svr_preds_test = svr.predict(X_test)
ensemble_preds_test = (linear_reg_preds_test + svr_preds_test) / 2
final_predictions = adaboost.predict(ensemble_preds_test.reshape(-1, 1))

mse = mean_squared_error(y_test, final_predictions)

r2 = r2_score(y_test, final_predictions)

print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R-squared (Accuracy): {r2:.2f}")
```

# K-means model

Finally, a K-Means model is taken into account. The elbow approach is used to calculate the ideal number of clusters. Clustering, on the other hand, is not the optimal strategy for this purpose because it is normally used to group data points based on similarities rather than forecasting final scores.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
X = df[features]
scaler = StandardScaler()
scaled_data = scaler.fit_transform(X)

sse=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k)
    kmeans.fit(scaled_data)
    sse.append(kmeans.inertia_)

plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```
